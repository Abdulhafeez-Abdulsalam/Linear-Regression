```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df= pd.read_csv('/content/sample_data/My_Data/CO2_emission.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.rename(columns={'Make':'Car_Brand'}, inplace=True)
df.head()
```

```python
X = df.iloc[:,3].values.reshape(-1,1)
X[:5]
```

```python

y = df.iloc[:,-1].values
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
my_model = LinearRegression()
my_model.fit(X_train, y_train)
y_pred = my_model.predict(X_test)
```

```python
y_pred
```

```python
error = y_pred - y_test
error
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red')
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
dseat['PRODUCTION DATE'] = pd.to_datetime(dseat['PRODUCTION DATE'],  format='%d/%m/%Y %H:%M', dayfirst=True)
dseat.columns = dseat.columns.str.strip()
 #Set the time column as the index for better plotting
dseat.set_index('PRODUCTION DATE', inplace=True)
dseat = dseat.sort_index()


# Plot Oil Production
plt.figure(figsize=(12, 6))
plt.plot(dseat.index, dseat['Oil Production (stb/day)'], label='Oil Production', color='red', linewidth=1)
plt.title('Oil Production Over Time', fontsize=14)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Oil Production (stb/day)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
